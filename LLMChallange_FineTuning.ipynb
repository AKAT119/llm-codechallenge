```python
from transformers import pipeline

def generate_pyomo_code(pipe, model_name):
    # Define the Pyomo template
    pyomo_template = """
from pyomo.environ import *
model = ConcreteModel()
model.x = Var(domain=NonNegativeReals)
model.y = Var(domain=NonNegativeReals)
model.obj = Objective(expr=2 * model.x + 3 * model.y, sense=maximize)

# Define constraints
model.constraint1 = Constraint(expr=model.x + 2 * model.y <= 8)
model.constraint2 = Constraint(expr=3 * model.x + 2 * model.y <= 12)

# Add the new constraint
{constraint}

# Solve the model
solver = SolverFactory('glpk')
result = solver.solve(model)
"""

    # Generate the constraint code without giving the mathematical formula directly
    inputs = "Please add constraint of 2 X + 6 Y > 30 to given pyomo code and rewrite entire code"
    outputs = pipe(inputs, max_length=100, num_return_sequences=1)
    constraint_code = outputs[0]['generated_text']

    # Insert the generated constraint into the template
    pyomo_code = pyomo_template.format(constraint=constraint_code)
    return pyomo_code

# Example usage
# Assuming model_starcoder and tokenizer_starcoder are already defined and loaded
pipe_starcoder = pipeline("text-generation", model=model_starcoder, tokenizer=tokenizer_starcoder)
pyomo_code_starcoder = generate_pyomo_code(pipe_starcoder, model_name="starcoder")
print("StarCoder Generated Code:\n", pyomo_code_starcoder)

# Assuming model_codelama and tokenizer_codelama are already defined and loaded
pipe_codelama = pipeline("text-generation", model=model_codelama, tokenizer=tokenizer_codelama)
pyomo_code_codelama = generate_pyomo_code(pipe_codelama, model_name="codelama")
print("CodeLlama Generated Code:\n", pyomo_code_codelama)
```

```python
from datasets import load_metric
import sacrebleu
from rouge_score import rouge_scorer

def evaluate_model(model, tokenizer, input_text, reference_texts):

    pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, device=torch.device('cpu'))
    generated_texts = pipe(input_text, max_length=100, num_return_sequences=1)
    generated_text = generated_texts[0]['generated_text']

    # BLEU Score
    bleu = sacrebleu.corpus_bleu([generated_text], [[ref] for ref in reference_texts])

    # ROUGE Score
    rouge = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)
    rouge_scores = [rouge.score(ref, generated_text) for ref in reference_texts]

    results = {
        "generated_text": generated_text,
        "bleu": bleu.score,
        "rouge": [
            {
                "rouge1": {
                    "precision": rouge['rouge1'].precision,
                    "recall": rouge['rouge1'].recall,
                    "fmeasure": rouge['rouge1'].fmeasure
                },
                "rougeL": {
                    "precision": rouge['rougeL'].precision,
                    "recall": rouge['rougeL'].recall,
                    "fmeasure": rouge['rougeL'].fmeasure
                }
            }
            for rouge in rouge_scores
        ]
    }
    return results


# Example usage
input_text = "Add the following constraint to a Pyomo code: 2 * model.x + 6 * model.y > 30"
reference_texts = ["model.constraint3 = Constraint(expr=2 * model.x + 6 * model.y > 30)"]

# # Evaluate StarCoder before fine-tuning
# results_starcoder_before = evaluate_model(model_starcoder, tokenizer_starcoder, input_text, reference_texts)
# print("StarCoder Before Fine-Tuning:\n", results_starcoder_before)

# # Evaluate CodeLlama before fine-tuning
# results_codelama_before = evaluate_model(model_codelama, tokenizer_codelama, input_text, reference_texts)
# print("CodeLlama Before Fine-Tuning:\n", results_codelama_before)
```

```python
from transformers import Trainer, TrainingArguments, pipeline
from datasets import load_dataset
from peft import get_peft_model, LoraConfig
import torch

# Load and sample 10% of the dataset
dataset = load_dataset("code_search_net", split="train", trust_remote_code=True)
sampled_dataset = dataset.shuffle(seed=42).select(range(int(0.1 * len(dataset))))
assert len(sampled_dataset) > 0, "Sampled dataset is empty!"

# Define the maximum length for tokenization
MAX_LENGTH = 512

# Set pad_token for the tokenizer if it doesn't exist
if tokenizer_starcoder.pad_token is None:
    tokenizer_starcoder.add_special_tokens({'pad_token': '[PAD]'})

# Tokenize the sampled dataset using the 'func_code_string' field
def tokenize_function(examples):
    return tokenizer_starcoder(examples["func_code_string"], padding="max_length", truncation=True, max_length=MAX_LENGTH)

tokenized_datasets = sampled_dataset.map(tokenize_function, batched=True)

# Verify tokenized dataset is not empty
assert len(tokenized_datasets) > 0, "Tokenized dataset is empty!"

# Split into train and eval datasets if necessary
train_size = int(0.8 * len(tokenized_datasets))
train_dataset = tokenized_datasets.select(range(train_size))
eval_dataset = tokenized_datasets.select(range(train_size, len(tokenized_datasets)))
```

```python
import torch

# Set up the device for StarCoder to use GPU
device_starcoder = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Set up the device for CodeLlama to use CPU
device_codelama = torch.device("cpu")
```

```python
from peft import get_peft_model, LoraConfig

# LoRA configuration for PEFT
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.1,
    target_modules=["q_proj", "v_proj"]
)

# Apply LoRA configuration to StarCoder
model_starcoder_lora = get_peft_model(model_starcoder, lora_config).to(device_starcoder)

# Apply LoRA configuration to CodeLlama
model_codelama_lora = get_peft_model(model_codelama, lora_config).to(device_codelama)
```

```python
from transformers import TrainingArguments

# Fine-tuning Setup for PEFT
training_args_lora_starcoder = TrainingArguments(
    output_dir="./results_lora_starcoder",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    num_train_epochs=1,
    weight_decay=0.01,
    push_to_hub=False,
    fp16=True,
)

training_args_lora_codelama = TrainingArguments(
    output_dir="./results_lora_codelama",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    num_train_epochs=1,
    weight_decay=0.01,
    push_to_hub=False,
    fp16=False,  # Ensure fp16 is disabled for CPU
)
```

```python
from transformers import Trainer

# Trainer for StarCoder with LoRA (using GPU)
trainer_starcoder_lora = Trainer(
    model=model_starcoder_lora,
    args=training_args_lora_starcoder,
    train_dataset=tokenized_datasets,
)
```

```python
# Trainer for CodeLlama with LoRA (using CPU)
trainer_codelama_lora = Trainer(
    model=model_codelama_lora,
    args=training_args_lora_codelama,
    train_dataset=tokenized_datasets,
)
```

```python
from peft import get_peft_model, PrefixTuningConfig

# Prefix Tuning configuration
prefix_tuning_config = PrefixTuningConfig(
    num_virtual_tokens=20,
    prefix_projection=True
)

# Apply Prefix Tuning configuration to StarCoder
model_starcoder_prefix = get_peft_model(model_starcoder, prefix_tuning_config)

# Apply Prefix Tuning configuration to CodeLlama
model_codelama_prefix = get_peft_model(model_codelama, prefix_tuning_config)

# Fine-tuning Setup for Prefix Tuning
training_args_prefix = TrainingArguments(
    output_dir="./results_prefix",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    num_train_epochs=1,
    weight_decay=0.01,
    push_to_hub=False,
    fp16=True,
)

# Trainer for StarCoder with Prefix Tuning
trainer_starcoder_prefix = Trainer(
    model=model_starcoder_prefix,
    args=training_args_prefix,
    train_dataset=tokenized_datasets,
)

# Trainer for CodeLlama with Prefix Tuning
trainer_codelama_prefix = Trainer(
    model=model_codelama_prefix,
    args=training_args_prefix,
    train_dataset=tokenized_datasets,
)

# Fine-tune the models with Prefix Tuning
trainer_starcoder_prefix.train()
trainer_codelama_prefix.train()
```
